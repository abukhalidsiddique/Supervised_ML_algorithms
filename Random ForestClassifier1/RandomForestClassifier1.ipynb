```python
import pandas as pd
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
df=pd.read_csv('team_combat.csv')
```

```python
x = df.First.values
```

```python
x
```

```python
y=df.Second.values
```

```python
y
```

```python
features= list(zip(x,y))    
```

```python
z=df.Winner.values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(features,z,test_size=.20)
```

```python
model=RandomForestClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
y_predict=model.predict(x_test)
```

```python
y_predict
```

```python
model.score(x_test,y_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model2=DecisionTreeClassifier()
```

```python
model2.fit(x_test,y_test)
```

```python
model2.score(x_test,y_test)
```

```python
y1_predict=model2.predict(x_test)
```

```python
accuracy_score(y_test, y_predict)
```

```python
precision_score(y_test, y_predict, average='weighted')
```
