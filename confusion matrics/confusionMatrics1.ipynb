```python
from sklearn.metrics import confusion_matrix
```

```python
y_actual = [0, 1, 0, 1, 1, 0, 0, 1, 1, 1]
```

```python
y_pred = [0, 1, 0, 1, 0, 1, 0, 0, 1, 1]
```

```python
cm = confusion_matrix(y_actual, y_pred)
```

```python
cm
```

```python
import seaborn as sns
```

```python
sns.heatmap(cm, annot=True, cmap='summer')
```
