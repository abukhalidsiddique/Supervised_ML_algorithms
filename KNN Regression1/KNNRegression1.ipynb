```python
df=pd.read_csv('book1.csv')
```

```python
x=df.Age.values     
```

```python
y=df.Height.values     
```

```python
z=df.Weight.values     
```

```python
features= list(zip(x,y))    
```

```python
from sklearn.neighbors import KNeighborsRegressor     
```

```python
model=KNeighborsRegressor(n_neighbors=3)
```

```python
model.fit(features,z)
```

```python
model.predict([[54, 6.2]])
```

```python
model.score(features,z)
```
